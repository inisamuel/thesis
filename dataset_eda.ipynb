```python
from utils import *
from aeon.datasets import load_from_tsv_file
import seaborn as sns
import pandas as pd
```

```python
dataset = "ArrowHead"
```

```python
X_train, y_train = load_from_tsv_file(get_dataset_path(dataset, "train"))
X_train = X_train.squeeze()
```

```python
df = pd.DataFrame(X_train).T
df.columns = y_train  # Assign column names as labels
df = df.reset_index().melt(id_vars='index', var_name='Label', value_name='Value')
df = df.rename(columns={'index': 'Time'})
df.head()
```

```python
sns.lineplot(data=df, x='Time', y='Value', hue='Label')
```
